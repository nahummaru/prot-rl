```python
from datasets import load_dataset
from Bio.Seq import Seq
```

```python
# Load the MegaScale dataset directly from Hugging Face
dataset = load_dataset("RosettaCommons/MegaScale", split="train", data_dir="dataset1")

# Convert to a pandas DataFrame
df = dataset.to_pandas()

df["protein_seq"] = df["dna_seq"].apply(lambda x: str(Seq(x).translate(to_stop=True)))

```

```python
# Display summary information
summary = {
    "num_rows": len(df),
    "columns": df.columns.tolist(),
    "missing_values": df.isnull().sum().to_dict(),
    "example_row": df.iloc[0].to_dict()
}

```
